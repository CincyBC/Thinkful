```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import math
```

```python
%matplotlib inline
red_df = pd.read_csv('red grapes.csv')
red_df.shape
print(red_df)
```

```python
plt.figure(figsize=(10, 5))
plt.plot(red_df['2016'], color = 'green')
keys = [x[0:5] for x in red_df['WINE GRAPES (RED):']]
x_pos = np.arange(len(keys))
plt.xticks(x_pos, keys)
plt.xlabel('Grapes')
plt.ylabel('Tons Produced')
plt.title('Grape Production 2016')

plt.show()
         
```

```python
plt.figure(figsize=(10, 5))
plt.plot(red_df['2016'], color = 'green')
plt.plot(red_df['2015'], color = 'blue')
keys = [x[0:5] for x in red_df['WINE GRAPES (RED):']]
x_pos = np.arange(len(keys))
plt.xticks(x_pos, keys)
plt.xlabel('Grapes')
plt.ylabel('Tons Produced')
plt.title('Grape Production 2015 and 2016')

plt.show()
         
```

```python
plt.scatter(x = red_df['2016'], y = red_df['2015'])
plt.xlabel('Grapes')
plt.ylabel('Tons Produced')
plt.title('Grape Production 2016')

plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.plot(red_df['2016'], color = 'green')
keys = [x[0:5] for x in red_df['WINE GRAPES (RED):']]
x_pos = np.arange(len(keys))
plt.xticks(x_pos, keys)
plt.xlabel('Grapes')
plt.ylabel('Tons Produced')
plt.title('Grape Production 2016')


plt.subplot(1, 2, 2)
plt.plot(red_df['2015'], color = 'red')
keys = [x[0:5] for x in red_df['WINE GRAPES (RED):']]
x_pos = np.arange(len(keys))
plt.xticks(x_pos, keys)
plt.xlabel('Grapes')
plt.ylabel('Tons Produced')
plt.title('Grape Production 2015')


plt.show()
```
