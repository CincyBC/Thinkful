```python
for key in keywords:
    # Note that we add spaces around the key so that we're getting the word,
    # not just pattern matching.
    original[str(key)] = original.message.str.contains(
        ' ' + str(key) + ' ',
        case=False
    )
original['impressions'] = (original['impressions'] == 1)
sns.heatmap(original.corr())
```

```python
data = original[keywords]
target = original['impressions']
# Instantiate our model and store it in a new variable.
bnb = BernoulliNB()

# Fit our model to the data.
bnb.fit(data, target)

# Classify, storing the result in a new variable.
y_pred = bnb.predict(data)
target_p = (target != y_pred).sum()
percent = round(((len(data) - target_p) / len(data)) * 100, 2)
# Display our results.
print("Accuracy rate out of a total {} points : {}%".format(
    data.shape[0], percent))
```

The above is based on the main keyword set I created used all of the data points as training. Let's see how it performs using the holdout method where we'll test it on a random sample of 20% of the data points. Really, we should have performed the keyword analysis on the first 80% training portion, but in this case, since I already did it for all 1000 data points, we're skipping that portion.

```python
#holdout
msk = np.random.rand(len(sms_raw)) < 0.8
train = sms_raw[msk]
test = sms_raw[~msk]

for key in keywords:
    # Note that we add spaces around the key so that we're getting the word,
    # not just pattern matching.
    test[str(key)] = test.message.str.contains(
        ' ' + str(key) + ' ',
        case=False
    )
test['impressions'] = (test['impressions'] == 1)

test_target = test['impressions']
test = test[keywords]
```

```python
# Instantiate our model and store it in a new variable.
bnb = BernoulliNB()

# Fit our model to the data.
bnb.fit(test, test_target)

# Classify, storing the result in a new variable.
y_pred = bnb.predict(test)
testtarget_p = (test_target != y_pred).sum()
testpercent = round(((len(test) - testtarget_p) / len(test)) * 100, 2)
# Display our results.
print("Accuracy rate out of a total {} points in test: {}%".format(
    test.shape[0], testpercent))
```
